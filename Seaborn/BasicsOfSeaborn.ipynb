```python
sns.scatterplot(x = "total_bill",y= "tip",data=data) # we are doing scatter point to dot sets
```

```python
sns.histplot(data["total_bill"]) # plotting histogram of data feature
```

```python
plt.hist(data["total_bill"]) # plotting histogram of data feature
```

```python
sns.kdeplot(data["total_bill"])
```

catplot

scatter plot

```python
sns.relplot(x="total_bill",y="tip",data=data) # relation plot b/w x and y
```

```python
data["time"].unique()
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",data=data)
#In Seaborn, the hue parameter is used to color-code data points or lines based on a categorical variable,
#  allowing for the visualization of multiple groups or categories within a single plot.
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",style="smoker",data=data)

# Relational plots are used for visualizing the statistical relationship between the data points
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",style="smoker",col="smoker",kind="line",data=data)
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",data=data,kind="scatter",style="smoker",row="smoker")
```

```python
data["day"].unique()
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",data=data,kind="line",style="smoker",col="day")
```

```python
sns.relplot(x="total_bill",y="tip",data=data)
plt.xlabel("x-axis total_bill")
plt.ylabel("y-axis tip")
plt.title("scatter plot")
```
